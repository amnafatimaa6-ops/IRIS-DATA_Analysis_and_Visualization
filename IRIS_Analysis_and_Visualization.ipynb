```python
import pandas as pd #for data analysis
#loading the data
df = pd.read_csv('/content/iris.csv')
#shape returns the number of rows and columns in the DataFrame
print("Shape of dataset:", df.shape)
#columns returns the list of column labels in the dataset
print("Column names:", df.columns)
#head() shows the first 5 rows by default, useful to quickly inspect the data
print("\nFirst 5 rows of the dataset:")
print(df.head())
```

```python
#info() shows column names, number of non-null values, and data types
print("\nDataset Information:")
df.info()
#describe() gives statistical measures like mean, std, min, max, etc. for numerical columns
print("\nSummary Statistics:")
print(df.describe())
```

The dataset contains 150 observations and 5 columns. Four columns (sepal_length, sepal_width, petal_length, petal_width) are numerical features stored as float64, while the species column is a categorical variable stored as an object. There are no missing values, as each column has 150 non-null entries.
From the statistical summary, the average sepal length is about 5.84 cm and the average sepal width is 3.05 cm. Petal measurements show greater variability, with an average petal length of 3.76 cm and average petal width of 1.20 cm. The standard deviation indicates that petal length varies more across samples compared to sepal measurements.
The minimum and maximum values show the range of measurements in the dataset. For example, sepal length ranges from 4.3 cm to 7.9 cm, while petal length ranges from 1.0 cm to 6.9 cm, suggesting noticeable diversity in flower sizes.


##EDA(Exploratory Data Analysis) & Visualization


```python
#for visualization
import seaborn as sns
import matplotlib.pyplot as plt
#count of each species
sns.countplot(x='species', data=df)
plt.title("Distribution of Iris Species")
plt.xlabel("Species")  #the plot shows that all species are fairly distribute in the data
plt.ylabel("Count")
plt.show()
```

```python
sns.pairplot(df, hue='species')
plt.show()
```

Insights:
Virginica flowers generally have the largest petals and long sepals.

Versicolor has medium-sized petal measurements.

Setosa is clearly distinguishable due to its small petals and tightly grouped measurements.

```python
plt.figure(figsize=(6,4))
corr = df.corr(numeric_only=True)
sns.heatmap(corr, annot=True, cmap='BuPu')
plt.title("Feature Correlation Heatmap")
plt.show()
```

Heatmap Insight

Sepal length: strongly correlated with petal length & width, weak with sepal width.

Sepal width: medium correlation with petals, weak with sepal length.

Petal length & width: strongly correlated with each other and sepal length, weak with sepal width.

Takeaway: Petal features are most informative for species classification.

```python
numerical_features = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width']

for i, feature in enumerate(numerical_features, 1):
    plt.subplot(2, 2, i)  # 2x2 grid
    sns.boxplot(x='species', y=feature, data=df, palette='Purples')
    plt.title(f"{feature} by Species")
plt.tight_layout()
plt.show()
```
